```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("seaborn")
```

```python
tickers=['IRFC.NS']
start='2023-01-01'
end=datetime.now()
```

```python
a=["^NSEI","TATACONSUM","TATAMOTORS","BAJAJ-AUTO","NESTLEIND","ADANIENT","BRITANNIA","HINDUNILVR","EICHERMOT","TATASTEEL","BAJFINANCE","ADANIPORTS","ULTRACEMCO","LTIM","BAJAJFINSV","MARUTI","WIPRO","HDFCBANK","LT","TCS","HINDALCO","ASIANPAINT","SUNPHARMA","SBILIFE","HDFCLIFE","UPL","HCLTECH","JSWSTEEL","GRASIM","ITC","M&M","BHARTIARTL","HEROMOTOCO","AXISBANK","RELIANCE","ICICIBANK","TITAN","DIVISLAB","CIPLA","KOTAKBANK","INDUSINDBK","APOLLOHOSP","DRREDDY","TECHM","POWERGRID","NTPC","COALINDIA","INFY","ONGC","SBIN","BPCL"]
tickers = [i+".NS" for i in a[1:]]
tickers.append("^NSEI")
start='2023-01-01'
end=datetime.now()
```

```python
import random
temp = tickers
rand_20_stocks=[]
for i in range(20):
  rand_20_stocks.append(temp.pop(random.choice(range(len(temp)))))
rand_20_stocks
```

```python
stocks = yf.download(rand_20_stocks,start=start, end=end)
```

```python
close = stocks.Close.copy()
```

```python
close.div(close.iloc[0]).plot(figsize=(12,8))
```

```python
ret = close.pct_change().dropna()
```

```python
ret.head()
```

```python
summary = ret.describe().T[['mean','std']]
summary['mean'] = summary['mean']*252
summary['std'] = summary['std']*np.sqrt(252)
```

```python
summary
```

```python
summary.plot.scatter(x='std',y='mean',figsize=(18,14),s=10,fontsize=15)
for i in summary.index:
  plt.annotate(i,xy=(summary.loc[i,'std']+0.004,summary.loc[i,'mean']+0.004),size=15)
plt.show()
```
